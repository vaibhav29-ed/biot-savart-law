```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad_vec
import plotly.graph_objects as go
import sympy as smp

phi = np.linspace(0, 2*np.pi, 100)

def l(phi):
    return (1 + (3/4) * np.sin(3*phi)) * np.array([np.cos(phi), np.sin(phi), np.zeros(len(phi))])

lx, ly, lz = l(phi)

plt.figure(figsize=(7, 7))
plt.plot(lx, ly)
plt.xlabel('$x/R$', fontsize=20)
plt.ylabel('$y/R$', fontsize=20)
plt.title('Wavy Current-Carrying Loop (XY plane)')
plt.axis('equal')
plt.grid(True)
plt.show()

t, x, y, z = smp.symbols('t x y z')
l = (1 + (3/4) * smp.sin(3*t)) * smp.Matrix([smp.cos(t), smp.sin(t), 0])
r = smp.Matrix([x, y, z])
sep = r - l
dl = smp.diff(l, t)
integrand = dl.cross(sep) / sep.norm()**3

dBxdt = smp.lambdify([t, x, y, z], integrand[0], 'numpy')
dBydt = smp.lambdify([t, x, y, z], integrand[1], 'numpy')
dBzdt = smp.lambdify([t, x, y, z], integrand[2], 'numpy')

def B(x, y, z):
    return np.array([
        quad_vec(dBxdt, 0, 2*np.pi, args=(x, y, z))[0],
        quad_vec(dBydt, 0, 2*np.pi, args=(x, y, z))[0],
        quad_vec(dBzdt, 0, 2*np.pi, args=(x, y, z))[0]
    ])

x = np.linspace(-2, 2, 10)
xv, yv, zv = np.meshgrid(x, x, x)

Bx = np.zeros_like(xv)
By = np.zeros_like(yv)
Bz = np.zeros_like(zv)

for i in range(xv.shape[0]):
    for j in range(xv.shape[1]):
        for k in range(xv.shape[2]):
            B_vec = B(xv[i, j, k], yv[i, j, k], zv[i, j, k])
            Bx[i, j, k], By[i, j, k], Bz[i, j, k] = B_vec

Bx = np.clip(Bx, -20, 20)
By = np.clip(By, -20, 20)
Bz = np.clip(Bz, -20, 20)

data = go.Cone(
    x=xv.ravel(), y=yv.ravel(), z=zv.ravel(),
    u=Bx.ravel(), v=By.ravel(), w=Bz.ravel(),
    colorscale='Inferno', sizemode="absolute", sizeref=20,
    showscale=True, colorbar=dict(title='|B|')
)

layout = go.Layout(
    title='3D Magnetic Field (BiotSavart Law)',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z',
        aspectratio=dict(x=1, y=1, z=1),
        camera_eye=dict(x=1.2, y=1.2, z=1.2)
    )
)

fig = go.Figure(data=[data], layout=layout)

lx, ly, lz = l(phi)
fig.add_trace(go.Scatter3d(
    x=lx, y=ly, z=lz,
    mode='lines',
    line=dict(color='green', width=6),
    name='Current Loop'
))

fig.show()
```
